```python
import pandas as pd
import pymysql

df = pd.read_excel("CovidDeaths.xlsx")
df
```

```python
df[~df.continent.isna()][['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'population']].sort_values(by=['location', 'date'])
```

```python
# Total Cases vs Total Deaths
temp = df[(~df.continent.isna())][['location', 'date', 'total_cases', 'total_deaths']]
temp['DeathPercentage'] = (temp.total_deaths / temp.total_cases)*100
temp
```

```python
# Total Cases vs Population
temp = df[(~df.continent.isna())][['location', 'date', 'population', 'total_cases']]
temp['PercentPopulationInfected'] = (temp.total_cases / temp.population)*100
temp
# temp.to_excel('PercentPopulationInfectedPerCountryWithDate.xlsx') 
```

```python
# Countries with Highest Infection Rate compared to Population
temp = df[['location', 'population', 'total_cases']]
temp1 = temp.groupby(['location', 'population']).agg({'total_cases':'max'}).rename(columns = {'total_cases':'HighestInfectionCount'})
temp1 = temp1.reset_index()
temp1['PercentPopulationInfected'] = temp1['HighestInfectionCount'] / temp1['population'] * 100
temp1
# temp1.to_excel('PercentPopulationInfectedPerCountry.xlsx')
```

```python
# Countries with Highest Death Count per Population
temp = df[(~df.continent.isna())][['location', 'total_deaths']]
temp1 = temp.groupby(['location']).agg({'total_deaths':'max'}).rename(columns = {'total_deaths':'TotalDeathCount'})
temp1.sort_values(['TotalDeathCount'], ascending = False).dropna().head(10)
# temp1.to_excel('DeathPerCountry.xlsx')
```

```python
# BREAKING THINGS DOWN BY CONTINENT
# Showing contintents with the highest death count per population
temp = df[(~df.continent.isna())][['continent', 'total_deaths']]
temp1 = temp.groupby(['continent']).agg({'total_deaths':'max'}).rename(columns = {'total_deaths':'TotalDeathCount'})
temp1 = temp1.sort_values(['TotalDeathCount'], ascending=False).reset_index()
temp1
# temp1.to_excel('DeathPerContinent.xlsx')
```

```python
# GLOBAL NUMBERS
temp = df[(~df.continent.isna())][['new_cases', 'new_deaths']]
temp1 = temp.agg({'new_cases':'sum', 'new_deaths':'sum'})
temp1['DeathPercentage'] = temp1['new_deaths'] / temp1['new_cases'] * 100
temp1 = temp1.to_frame().transpose()
temp1
temp1.to_excel('dataset/DeathGlobal.xlsx')
```

```python
df1 = pd.read_excel("CovidVaccinations.xlsx")
df1
```

```python
df_total = df.merge(df1, left_on=['location', 'date'], right_on=['location', 'date'])
df_total
```

```python
# Total Population vs Vaccinations
# Shows Percentage of Population that has recieved at least one Covid Vaccine
temp = df_total[(~df_total.continent_x.isna())][['continent_x', 'location', 'date', 'population', 'new_vaccinations_y']].sort_values(['date', 'population']).reset_index()
temp = temp.groupby(['location', 'population']).agg({'new_vaccinations_y':'sum'}).reset_index()
temp
```

```python
temp = df_total[(~df_total.continent_x.isna())][['continent_x', 'location', 'date', 'population', 'new_vaccinations_y']].sort_values(['location','date']).reset_index()
temp['rolling'] = temp.groupby(['location'])['new_vaccinations_y'].cumsum()
temp.fillna(0, inplace=True)
temp
temp.to_excel('DeathPerCountry.xlsx')
```
